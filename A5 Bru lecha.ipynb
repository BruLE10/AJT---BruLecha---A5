```python
class Song:
    def __init__(self, titol, autor, durada):
        self.titol = titol
        self.autor = autor
        self.durada = durada 

    def info(self):
        return f"{self.titol} - {self.autor} ({self.durada} segons)"


```

```python
class playlist:
    def __init__(self, list_song):
        self.list_song = list_song
        list_song = []

    def afegir_canco (self, canco):
        self.list_song.append(canco)

    def durada_total (self):
        total=0
        for n in self.list_song:
            total += n.durada
        return total
             
    def mostrar_playlist (self):
        for i in self.list_song:
            print (i.info())
   
    def canco_mes_llarga (self):
        canco_mes_llarga = self.list_song[0]
        for t in self.list_song:
            if t.durada > canco_mes_llarga.durada:
                canco_mes_llarga = t
        return canco_mes_llarga.info()

```

```python
canco1 = Song("Bebe", "Anuel", 223)
canco2 = Song("Mala", "Anuel", 203)

playlistanuel = playlist ([canco1, canco2])

canco3= Song("Amamnece", "Anuel", 254)

playlistanuel.afegir_canco (canco3)
```

```python
print (playlistanuel.durada_total())
```

```python
print (playlistanuel.canco_mes_llarga())
```

```python
playlistanuel.mostrar_playlist()
```

```python
class Personatge:
    def __init__(self, nom, vida, forca, nivell):
        self.nom = nom
        self.vida = vida
        self.forca = forca
        self.nivell = nivell

    def atacar(self, enemic):
        print(f"{self.nom} ataca a {enemic.nom} amb una força de {self.forca}!")
        enemic.vida -= self.forca
        if enemic.vida < 0:
            enemic.vida = 0

    def pujar_nivell(self):
        self.nivell += 1
        self.forca += 2
        self.vida += 5
        print(f"{self.nom} ha pujat al nivell {self.nivell}! (Vida +5, Força +2)")

    def esta_viu(self):
        return self.vida > 0

    def __str__(self):
        return f"Personatge: {self.nom}, Vida: {self.vida}, Força: {self.forca}, Nivell: {self.nivell}"

    def info(self):
        print(self.__str__())


import time

p1 = Personatge("Aragorn", 1000, 250, 5)
p2 = Personatge("Legolas", 1000, 225, 5)

print("INICI DEL COMBAT")
print(p1)
print(p2)

torn = 1
while p1.esta_viu() and p2.esta_viu():
    print(f"\nTORN {torn}:")
    
    p1.atacar(p2)
    if not p2.esta_viu():
        break
        
    p2.atacar(p1)
    
    print(f"Vida: {p1.nom}({p1.vida}) vs {p2.nom}({p2.vida})")
    torn += 1
    time.sleep(0.5)

if p1.esta_viu():
    guanyador = p1
    print(f"\nGUANYADOR: {p1.nom.upper()}")
else:
    guanyador = p2
    print(f"\nGUANYADOR:{p2.nom.upper()}")

guanyador.pujar_nivell()
guanyador.info()
```
